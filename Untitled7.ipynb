```python
def get_param(df):
  param = {}
  for i in df.columns[df.isnull().any()]:
    if df[i].dtype == 'float64' or df[i].dtype == 'int32' or df[i].dtype == 'int64':
      strategy = 'mean'
    else:
      strategy = 'most_frequent'

    missing_value = df[i][df[i].isnull()].values[0]
    param[i] = {'missing_value':missing_value , 'strategy':strategy}
  return param
get_param(df)
```

```python
parameters = get_param(df);
for col , param in parameters.items():
  print(col)
```

```python
for col , param in parameters.items():
  missing_values = param['missing_value']
  # print(param['strategy'])
  strategy = param['strategy']
  imp = SimpleImputer( missing_values = missing_values , strategy=strategy)
  A = pd.DataFrame()
  A = imp.fit_transform(df[[col]].values)
  A = pd.DataFrame(A)
  df[col] = A
```
